# ğŸ“ Definisi *Norm* Vektor

**Norm** dari sebuah vektor adalah ukuran panjang atau magnitudo vektor tersebut dalam ruang vektor. Norm dilambangkan sebagai:
- $\|\vec{v}\|$ (notasi standar)
- $|\vec{v}|$ (alternatif notasi)

## Formula Norm Euclidean

Untuk vektor $\vec{v} = (v_1, v_2, \ldots, v_n) \in \mathbb{R}^n$, norm Euclidean (L2-norm) dihitung sebagai:

$$\|\vec{v}\| = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2} = \sqrt{\sum_{i=1}^{n} v_i^2}$$

## Contoh Perhitungan

**Contoh 1:** Vektor 2D
- Untuk $\vec{v} = (3, 4)$
- $\|\vec{v}\| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5$

**Contoh 2:** Vektor 3D
- Untuk $\vec{u} = (1, 2, 2)$
- $\|\vec{u}\| = \sqrt{1^2 + 2^2 + 2^2} = \sqrt{1 + 4 + 4} = \sqrt{9} = 3$

## Sifat-sifat Norm

1. **Non-negatif:** $\|\vec{v}\| \geq 0$ untuk semua $\vec{v}$
2. **Definiteness:** $\|\vec{v}\| = 0$ jika dan hanya jika $\vec{v} = \vec{0}$
3. **Homogenitas:** $\|c\vec{v}\| = |c| \cdot \|\vec{v}\|$ untuk skalar $c$
4. **Ketaksamaan segitiga:** $\|\vec{u} + \vec{v}\| \leq \|\vec{u}\| + \|\vec{v}\|$

## Interpretasi Geometris

Norm vektor memberikan jarak dari titik asal (origin) ke titik yang direpresentasikan oleh vektor tersebut dalam ruang koordinat. Ini adalah generalisasi dari teorema Pythagoras ke dimensi yang lebih tinggi.

# Hukum Sinus dan Kosinus

Hukum sinus dan kosinus digunakan untuk menyelesaikan nilai-nilai yang tidak diketahui dalam **segitiga sembarang** (bukan segitiga siku-siku).

---

## ğŸ”· Hukum Sinus

Digunakan saat kita mengetahui **dua sudut dan satu sisi (AAS atau ASA)** atau **dua sisi dan satu sudut yang berlawanan (SSA)**.

$$\frac{a}{\sin A} = \frac{b}{\sin B} = \frac{c}{\sin C}$$

**Keterangan:**
- a, b, c = panjang sisi-sisi segitiga
- A, B, C = sudut yang berhadapan dengan sisi a, b, dan c

### Kapan Menggunakan Hukum Sinus:
- **AAS (Angle-Angle-Side)**: Diketahui dua sudut dan satu sisi
- **ASA (Angle-Side-Angle)**: Diketahui dua sudut dan sisi yang diapit
- **SSA (Side-Side-Angle)**: Diketahui dua sisi dan satu sudut yang berlawanan dengan salah satu sisi

---

## ğŸ”· Hukum Kosinus

Digunakan saat kita mengetahui:
- **Tiga sisi segitiga (SSS)**, atau
- **Dua sisi dan sudut yang diapit (SAS)**

$$c^2 = a^2 + b^2 - 2ab\cos C$$

**Keterangan:**
- a, b, c = panjang sisi-sisi segitiga
- C = sudut yang berada di antara sisi a dan b

### Variasi Rumus Hukum Kosinus:
$$a^2 = b^2 + c^2 - 2bc\cos A$$
$$b^2 = a^2 + c^2 - 2ac\cos B$$
$$c^2 = a^2 + b^2 - 2ab\cos C$$

### Untuk Mencari Sudut:
$$\cos A = \frac{b^2 + c^2 - a^2}{2bc}$$
$$\cos B = \frac{a^2 + c^2 - b^2}{2ac}$$
$$\cos C = \frac{a^2 + b^2 - c^2}{2ab}$$

### Kapan Menggunakan Hukum Kosinus:
- **SSS (Side-Side-Side)**: Diketahui ketiga sisi segitiga
- **SAS (Side-Angle-Side)**: Diketahui dua sisi dan sudut yang diapit

---
