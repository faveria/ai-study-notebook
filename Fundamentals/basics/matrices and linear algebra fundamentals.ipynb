# ğŸ“ Definisi *Norm* Vektor

**Norm** dari sebuah vektor adalah ukuran panjang atau magnitudo vektor tersebut dalam ruang vektor. Norm dilambangkan sebagai:
- $\|\vec{v}\|$ (notasi standar)
- $|\vec{v}|$ (alternatif notasi)

## Formula Norm Euclidean

Untuk vektor $\vec{v} = (v_1, v_2, \ldots, v_n) \in \mathbb{R}^n$, norm Euclidean (L2-norm) dihitung sebagai:

$$\|\vec{v}\| = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2} = \sqrt{\sum_{i=1}^{n} v_i^2}$$

## Contoh Perhitungan

**Contoh 1:** Vektor 2D
- Untuk $\vec{v} = (3, 4)$
- $\|\vec{v}\| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5$

**Contoh 2:** Vektor 3D
- Untuk $\vec{u} = (1, 2, 2)$
- $\|\vec{u}\| = \sqrt{1^2 + 2^2 + 2^2} = \sqrt{1 + 4 + 4} = \sqrt{9} = 3$

## Sifat-sifat Norm

1. **Non-negatif:** $\|\vec{v}\| \geq 0$ untuk semua $\vec{v}$
2. **Definiteness:** $\|\vec{v}\| = 0$ jika dan hanya jika $\vec{v} = \vec{0}$
3. **Homogenitas:** $\|c\vec{v}\| = |c| \cdot \|\vec{v}\|$ untuk skalar $c$
4. **Ketaksamaan segitiga:** $\|\vec{u} + \vec{v}\| \leq \|\vec{u}\| + \|\vec{v}\|$
